```python
import requests
from bs4 import BeautifulSoup

url= "http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/"
info= {"h4":("propPrice",), "span":("propAddressCollapse", "infoBed","infoSqFt", "infoValueFullBath", "infoValueHalfBath")}

headers= {
"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.75 Safari/537.36"
}
r= requests.get(url, stream=True, headers=headers)
content= r.content

soup=BeautifulSoup(content, "html.parser")
all= soup.find_all("div",{"class":"propertyRow"})

for items in all:
    for keys in info:
        for values in info[keys]:
            try:
                occurence= len(items.find_all(keys,{"class":values}))
                i = 0
                if occurence>1:
                    while i <= occurence:
                        print(items.find_all(keys,{"class":values})[i].text.strip())
                        i= i+1
                else:
                    print(items.find(keys,{"class":values}).text.strip())
            except:
                print(None)
    print(" ")
```
